```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
f = open("C:/Navneet/College/Fall21/NLP/Song-Genre-Identifier/DataProcessor/dataset/processed/songList.json")
data = json.load(f)
f.close()
df = pd.json_normalize(data['songs'])
df.head()
```

```python
df['Lyrics'][0][0][1]
```

```python
def totalWords(arr):
    tot = 0
    freq = {}
    print(arr)
    for x in arr:
        tot = tot + len(x)
        for y in x:
            if(y in freq):
                freq[y]=freq[y]+1
            else:
                freq[y]=1
    WordMax = max(zip(freq.values(), freq.keys()))[1]
    WordFreq = max(zip(freq.values(), freq.keys()))[0]
    return tot, len(freq), WordMax, WordFreq
```

```python
lengthArray = []
uniqueArray = []
mostFrequentWord = []
freqOfMFW = []
for index,row in df.iterrows():
    n,freq,mfw,fmfw = totalWords(row.Lyrics)
    lengthArray.append(n)
    uniqueArray.append(freq)
    mostFrequentWord.append(mfw)
    freqOfMFW.append(fmfw)
df['TotalWords'] = pd.Series(lengthArray, index=df.index)
df['UniqueWords'] = pd.Series(uniqueArray, index = df.index)
df['MostFrequentWords'] = pd.Series(mostFrequentWord, index=df.index)
df['FrequencyOfMFW'] = pd.Series(freqOfMFW, index=df.index)
```

```python
def sentenceProcessing(arr):
    freq = {}
    total = 0
    for x in arr:
        sent = ' '.join(x)
        if sent in freq:
            freq[sent] = freq[sent]+1
        else:
            freq[sent] = 1
        total = total+1
    sentMax = max(zip(freq.values(), freq.keys()))[1]
    sentCount = max(zip(freq.values(), freq.keys()))[0]
    return total, sentMax, sentCount
```

```python
sentM = []
sentC = []
totalS = []
for index,row in df.iterrows():
    total, sentMax, sentCount = sentenceProcessing(row['Lyrics'])
    sentM.append(sentMax)
    sentC.append(sentCount)
    totalS.append(total)
df['TotalSentences'] = pd.Series(totalS, index = df.index)
df['MostFrequentSentence'] = pd.Series(sentM, index = df.index)
df['CountMostSentence'] = pd.Series(sentC, index = df.index)
```
